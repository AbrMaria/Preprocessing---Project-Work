6. С помощью статистических методов проверить взаимосвязи между переменными.

```python
stats.ttest_ind(data_worst['concave_points_worst'], data_worst['diagnosis'])
```

```python
stats.ttest_rel(data_worst['area_worst'], data_worst['diagnosis'])
```

```python
est = smf.ols('diagnosis ~ area_worst + radius_worst + perimeter_worst + concave_points_worst', data_worst).fit()
est_res = est.summary()
est_res.tables[1]
```

```python
#все выбранные показатели, кроме perimeter_worst, имеют маленький  p-value. 
#Т.е. между perimeter_worst и diagnosis нет линейной зависимости и, следовательно, от этого атрибута можно избавиться.
```

```python
three_x_lm = smf.ols('diagnosis ~ area_worst + radius_worst + concave_points_worst', data_worst).fit()
rss = np.sum(three_x_lm.resid ** 2)
print("RSS:", rss)
print("RSE:", np.sqrt(rss / (data_worst.shape[0] - 3 - 1)))
print("R^2:", three_x_lm.rsquared)
```

```python
two_x_lm = smf.ols('diagnosis ~ area_worst + radius_worst', data_worst).fit()
rss = np.sum(two_x_lm.resid ** 2)
print("RSS:", rss)
print("RSE:", np.sqrt(rss / (data_worst.shape[0] - 2 - 1)))
print("R^2:", two_x_lm.rsquared)
```
